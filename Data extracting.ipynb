```python
import rasterio
import numpy as np
import pandas as pd
import os

# Define the folder path
folder_path = r'C:\Users\user\Downloads\EO_Browser'

# List of specific filenames you want to process: B04,B08,B11,B12 bands
specific_files = ['B04.tiff', 'B08.tiff', 'B11.tiff', 'B12.tiff']

# Create the paths to all the files 
image_file_B04 = os.path.join(folder_path, specific_files[0])
image_file_B08 = os.path.join(folder_path, specific_files[1])
image_file_B11 = os.path.join(folder_path, specific_files[2])
image_file_B12 = os.path.join(folder_path, specific_files[3])

rows = []
cols = []

#### Get some code for tilting the data

with rasterio.open(image_file_B04) as src:
    # Read the image data into an array
    image_data = src.read()
    # Get the metadata of the image
    meta = src.meta
    # Extract the width and height of the image
    width, height = src.width, src.height



## B04 Band
#====================

b4_values = []

# Open the Sentinel-2 image
with rasterio.open(image_file_B04) as src:
    # Read the image data into an array
    image_data = src.read()
    # Get the metadata of the image
    meta = src.meta
    # Extract the width and height of the image
    width, height = src.width, src.height
    # Get the list of band names
    band_names = src.descriptions

# Iterate over each band
for band_index, band_name in enumerate(band_names, start=1):
    # Read the band data
    band_data = image_data[band_index - 1]
    
    # Iterate over each pixel in the band
    for row in range(height):
        for col in range(width):
            # Append the pixel location and value to the lists
            rows.append(row)
            cols.append(col)
            b4_values.append(band_data[row, col])

## B08 Band
#====================

b8_values = []

# Open the Sentinel-2 image
with rasterio.open(image_file_B08) as src:
    # Read the image data into an array
    image_data = src.read()
    # Get the metadata of the image
    meta = src.meta
    # Extract the width and height of the image
    width, height = src.width, src.height
    # Get the list of band names
    band_names = src.descriptions

# Iterate over each band
for band_index, band_name in enumerate(band_names, start=1):
    # Read the band data
    band_data = image_data[band_index - 1]
    
    # Iterate over each pixel in the band
    for row in range(height):
        for col in range(width):
            # Append the pixel location and value to the lists
            b8_values.append(band_data[row, col])
            
## B11 Band
#====================

b11_values = []

# Open the Sentinel-2 image
with rasterio.open(image_file_B11) as src:
    # Read the image data into an array
    image_data = src.read()
    # Get the metadata of the image
    meta = src.meta
    # Extract the width and height of the image
    width, height = src.width, src.height
    # Get the list of band names
    band_names = src.descriptions

# Iterate over each band
for band_index, band_name in enumerate(band_names, start=1):
    # Read the band data
    band_data = image_data[band_index - 1]
    
    # Iterate over each pixel in the band
    for row in range(height):
        for col in range(width):
            # Append the pixel location and value to the lists
            b11_values.append(band_data[row, col])

            
print('yes')            
## B12 Band
#====================

b12_values = []

# Open the Sentinel-2 image
with rasterio.open(image_file_B12) as src:
    # Read the image data into an array
    image_data = src.read()
    # Get the metadata of the image
    meta = src.meta
    # Extract the width and height of the image
    width, height = src.width, src.height
    # Get the list of band names
    band_names = src.descriptions

# Iterate over each band
for band_index, band_name in enumerate(band_names, start=1):
    # Read the band data
    band_data = image_data[band_index - 1]
    
    # Iterate over each pixel in the band
    for row in range(height):
        for col in range(width):
            # Append the pixel location and value to the lists
            b12_values.append(band_data[row, col])

print('yes')

b4_values = np.array(b4_values)
b8_values = np.array(b8_values)
b11_values = np.array(b11_values)
b12_values = np.array(b12_values)
            
NDWI = (b8_values-b11_values)/(b8_values+b11_values)
NDVI = (b8_values-b4_values)/(b8_values+b4_values)
BNR = (b8_values-b12_values)/(b8_values+b12_values)

print(len(NDWI))
print(len(NDVI))
print(len(BNR))
print(len(rows))
print(len(cols))
print(len(b11_values))
print(len(b12_values))
            
# Create a DataFrame with the extracted data
df = pd.DataFrame({
    'row': rows,
    'col': cols,
    'NDWI': NDWI,
    'NDVI': NDVI, 
    'BNR': BNR,
    'SWRI_1': b11_values,
    'SWRI_2': b12_values
})

# Save the DataFrame to a CSV file
csv_file = r'C:\Users\user\Desktop\DataAll.csv'
df.to_csv(csv_file, index=False)

print(f"Data extracted and saved to {csv_file}")
```
